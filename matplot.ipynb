```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
years = np.arange(1949, 2017, 1)
print(years)
```

```python
data = np.random.random(2017-1949) * 10
print(data)
```

```python
plt.plot()
plt.show()
```

```python
plt.plot(years, data)
plt.show()
```

```python
plt.plot(years, data)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.plot(years, data)
plt.xticks(rotation=45)
plt.xlabel('$year$')
plt.ylabel('$data$')
plt.title('$the\ data\ of\ each\ year$')
plt.show()
```

```python
# 子图结构
```

```python
fig = plt.figure()
pic1 = fig.add_subplot(2, 3, 1)
pic2 = fig.add_subplot(2, 3, 2)
pic3 = fig.add_subplot(2, 3, 3)
pic6 = fig.add_subplot(2, 3, 6)
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))
pic1 = fig.add_subplot(2, 3, 1)
pic1.plot(years, data)
pic2 = fig.add_subplot(2, 3, 2)
pic2.plot(years, data, 'k--', color='g')
pic3 = fig.add_subplot(2, 3, 3)
pic3.plot(years, data, c='blue')
pic6 = fig.add_subplot(2, 3, 6)
pic6.plot(years, data, color='blue', linestyle='--')
plt.show()
```

```python
x = np.linspace(-3.0, 3.0, 50)
y1 = x
y2 = x ** 2
y3 = x ** 3
plt.figure(figsize=(15,8))
plt.plot(x, y1, label='1')
plt.plot(x, y2, label='2')
plt.plot(x, y3, label='3')
plt.legend(loc='best')
plt.show()
```

```python
# 条形图
```

```python
size = 2017 - 1949
bar_positions = np.arange(size) + 1
plt.figure(figsize=(15, 8))
#  plt.bar(bar_positions, data, 0.5)
plt.bar(years, data, 0.5)
plt.plot(years, data, c='red')
plt.show()
```

```python
size = 2017 - 1949
bar_positions = np.arange(size) + 1
plt.figure(figsize=(15, 8))
#  plt.bar(bar_positions, data, 0.5)
plt.barh(years, data, 0.5)
plt.show()
```

```python
fig, ax = plt.subplots()
fig.figsize = (1, 20)
x = np.linspace(1, 10, 50)
y = np.random.random(50) * 5
print(x)
print(y)
ax.scatter(x, y, c='red')
plt.show()
```
